```python
import string
import re

words = text.split()
punctuation = re.compile('[%s]' % re.escape(string.punctuation))
strip = [punctuation.sub('',w) for w in words]
print(strip[:100])
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
text = "hello everyone i am ayush kumar !"
print(text)
ans = nlp(text)
for token in ans:
  print(token.text,end=' | ')
```

```python
nlp = spacy.load('en_core_web_sm')
string = nlp(u"spaCy excels at large-scale information extraction tasks. It's written from the ground up in carefully memory-managed Cython. If your application needs to process entire web dumps, spaCy is the library you want to be using.")
for token in string:
  print(token.text,end=' | ')
print('\n ')
for var in string.ents:
  print(var.text,var.label_)
```

```python
doc = nlp(u"India is buliding new Sri Ram Temple in Ayodhya.")
for part in string.noun_chunks:
  print(part)
for ent in doc.ents:
   print(ent.text, ent.label_)
len(string)
```
