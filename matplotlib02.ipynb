```python
import matplotlib.pyplot as plt
import numpy as np
import sys

number = np.array([1,3,4,6,7,8])
frequency = np.array([23,45,67,34,12,80])
plt.xlabel("integer numbers")
plt.ylabel("frequency")
plt.plot(number,frequency)
#plt.grid()    # if not specify than plt.grid() method will show x and y axis both axis grid for specific axis we can mention it
plt.grid(axis='x') # specifies only x axis
plt.grid(axis='y',linestyle='--',linewidth='1.4')    # specifies only y axis

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sys

income = np.array([100,200,500,700,350])
sales = np.array([20,34,65,32,160])
plt.title("sales vs income")
plt.subplot(1,2,1)
plt.plot(sales,income)
plt.xlabel("sales")
plt.ylabel("income")
plt.grid()

fruits = np.array(["apple","peanut","pine apple"])
price = np.array([120,300,450])
plt.subplot(1,2,2)
plt.plot(fruits,price)
plt.xlabel("fruits")
plt.ylabel("price")
plt.grid()
#  plt.clf() it will remove all the plot drawn above this line of code

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sys

age = np.array([10,12,31,23,43,14])
iq = np.array([110,100,87,143,80,130])
plt.xlabel("age of students")
plt.ylabel("iq of students")
plt.subplot(1,2,1)
plt.plot(age,iq,marker="*")
plt.grid()

plt.subplot(1,2,2)
colors = np.array([0, 10, 20, 30, 55, 60])
plt.xlabel("age of students")
plt.ylabel("iq of students")
plt.grid()
plt.scatter(age,iq,c=colors,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
import numpy as np
import sys
import matplotlib.pyplot as plt

integer = np.array([1,2,3,4,5])
frequency = np.array([10,12,34,45,11])
plt.xlabel("integer numbers")
plt.ylabel("frequency of numbers")
plt.bar(integer,frequency,width =0.1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import sys

heat = np.array([20,40,60,80,100])
vibration = np.array([35,100,135,180,240])
plt.xlabel("heat")
plt.ylabel("vibration")
plt.bar(heat,vibration,width=0.5)
#plt.barh(heat,vibration,height=0.6)
plt.show()
```
